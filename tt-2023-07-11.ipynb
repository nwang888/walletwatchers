Import Packages / Datasets

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
global_temps = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2023/2023-07-11/global_temps.csv')
nh_temps = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2023/2023-07-11/nh_temps.csv')
sh_temps = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2023/2023-07-11/sh_temps.csv')
zonann_temps = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2023/2023-07-11/zonann_temps.csv')
```

```python
global_temps['Average'] = global_temps.iloc[:,1:13].mean(axis=1)
gtemps = plt.figure()
axes = gtemps.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(global_temps['Year'], global_temps['Average'])
axes.set_xlabel('Year')
axes.set_ylabel('Average Temperature')
axes.set_title('Average Global Temperature Over Time')
```

```python
nh_temps['Average'] = nh_temps.iloc[:,1:13].mean(axis=1)
nhtemps = plt.figure()
axes = nhtemps.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(nh_temps['Year'], nh_temps['Average'])
axes.set_xlabel('Year')
axes.set_ylabel('Average Temperature')
axes.set_title('Average NH Temperature Over Time')
```

```python
sh_temps['Average'] = sh_temps.iloc[:,1:13].mean(axis=1)
shtemps = plt.figure()
axes = shtemps.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(sh_temps['Year'], sh_temps['Average'])
axes.set_xlabel('Year')
axes.set_ylabel('Average Temperature')
axes.set_title('Average SH Temperature Over Time')
```
